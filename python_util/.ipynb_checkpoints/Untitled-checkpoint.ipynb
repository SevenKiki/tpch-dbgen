```python
def process_strings():
    n = int(input("请输入字符串的行数："))
    strings = []

    for i in range(n):
        string = input("请输入第{}行的字符串：".format(i+1))
        substring = string.split("|")[0]
        strings.append(substring)

    print("处理结果：")
    for string in strings:
        print(string)


process_strings()
```
